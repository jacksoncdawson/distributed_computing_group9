```python
"""Grabs a sample of <sample_size> rows from the given full dataset, drops it in data_samples in the same .jsonl format. 
   Used for testing Spark jobs before pushing to GCS.
"""

import random
from pathlib import Path

category_name = "Video_Games"
sample_size = 50000 

input_path = Path(f"../../data_full/{category_name}.jsonl")
output_path = Path(f"../../data_samples/{category_name}_SAMPLE.jsonl")

# Count lines
with input_path.open("r", encoding="utf-8") as f:
    total_lines = sum(1 for _ in f)
    
if sample_size > total_lines:
    raise ValueError(f"Requested {sample_size} lines, but file has {total_lines}")


# Choose random unique line indices to keep
indices = set(random.sample(range(total_lines), sample_size))


# Write selected lines
with input_path.open("r", encoding="utf-8") as fin, \
     output_path.open("w", encoding="utf-8") as fout:
    for idx, line in enumerate(fin):
        if idx in indices:
            fout.write(line)

print(f"Wrote {sample_size} random lines to {output_path}")
```
